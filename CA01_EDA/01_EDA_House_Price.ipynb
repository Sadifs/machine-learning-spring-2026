```python
##PRE-PROCESSING##
```

```python
#Dropping Columns with High Percentage (About 50% or More) of Missing Values
df.drop(columns=['PoolQC', 'MiscFeature', 'Alley', 'Fence'], inplace=True)
```

```python
#Filling Missing Categorical Values Based on Assumption that They 
#FireplaceQu
df['FireplaceQu'] = df['FireplaceQu'].fillna('None')

#MasVnrType
df['MasVnrType'] = df['MasVnrType'].fillna('None')

#Garage
for col in ['GarageType', 'GarageFinish', 'GarageQual', 'GarageCond']:
    df[col] = df[col].fillna('None')
df['GarageYrBlt'] = df['GarageYrBlt'].fillna(0)

#Basement
for col in ['BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2']:
    df[col] = df[col].fillna('None')
```

```python
#Missing Numerical Values
#MasVnrArea
df['MasVnrArea'] = df['MasVnrArea'].fillna(0)
```

```python
#Imputing Very Low % Missing Values with Median and Mode
#LotFrontage
df['LotFrontage'] = df['LotFrontage'].fillna(df['LotFrontage'].median())

#Electrical
df['Electrical'] = df['Electrical'].fillna(df['Electrical'].mode()[0])
```

```python
#Confirming All Missing Values Got Addressed
df.isnull().sum().sum() 
```

```python
#Applying Log Transformation to Sale Price to Reduce Skewness While Retaining All Observations
##AI Assitance: https://chatgpt.com/share/69813076-8550-8011-9090-019d4a670618
df['SalePrice'] = np.log1p(df['SalePrice'])
```

```python
##Pre-Processing Notes##

##Categorical features where missing values could indicate the absence of a feature were filled with None
##Similar numerical features were filled with 0
##Features with very low missingness were imputed using median or mode to preserve the overall distribution

##Outliers in were not removed, as they likely represent valid high-value properties rather than data errors
##Instead, a log transformation was applied to SalePrice to reduce right skewness.

##Categorical encoding and scaling were deferred since the focus of CA01 is EDA and data quality assessment, not model training.
```

```python
##POST PROCESSING##
```

```python
#Visualizing Key Correlatioms with a Heatmap to Understand Collinearity
key_corr = corr_matrix[corr_matrix > 0.6].index

plt.figure(figsize=(12, 8))
sns.heatmap(df[key_corr].corr(numeric_only=True),cmap='RdYlGn',annot=True)
plt.title('Top Correlated Features with SalePrice')
plt.show()

##Used to identify multicollinearity among features and support feature selection decisions
##Highly correlated features (GarageArea & GarageCars) were flagged for removal to reduce redundancy
```

```python
#Collinearity-Based Feature Selection (Using Heatmap) to Reduce Redunancy 
df.drop(columns=['GarageArea'], inplace=True)
```

```python
##Post Processing Notes & Conclusions##

##Based on the observed correlations, features with strong linear relationships were removed

##Overall, this exploratory analysis provided a clear understanding of the dataset’s structure, quality issues & feature relationships
##The dataset is now clean & ready for use in subsequent modeling
```
